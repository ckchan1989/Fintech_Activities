```python
# Initial imports
import pandas as pd
from dotenv import load_dotenv
import os
from newsapi import NewsApiClient

```

```python
# Load env variables and read the News API Key enviroment variable
load_dotenv()
api_key = os.getenv("NEWS_API_KEY")
type(api_key)
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=api_key)

```

```python
# Fetch top news articles about oil in Canada 
#Note: If there are no headlines, code in upcoming cells will throw errors.
# If this happens, use the `get_everything()` function instead.

au_top_headlines = newsapi.get_top_headlines(q="australia", language="en", country="au")
au_top_headlines
```

```python
au_all_headlines = newsapi.get_everything(q="australia", language="en")

au_all_headlines
```

```python
# Print total articles
print(f"Total articles about australia: {au_top_headlines['totalResults']}")

# Show sample article
au_top_headlines["articles"][0]

```

```python
# Transform the response dictionary to a DataFrame
australia_df = pd.DataFrame.from_dict(au_top_headlines["articles"])

australia_df.head()

```

```python
libra_top_headlines = newsapi.get_top_headlines(q="facebook libra", language="en")
libra_top_headlines
```

```python
# Fetch all the news about Facebook Libra
libra_headlines = newsapi.get_everything(
    q="facebook libra", language="en", sort_by="relevancy"
)

# Print total articles
print(f"Total articles about Facebook Libra: {libra_headlines['totalResults']}")

# Show sample article
libra_headlines["articles"][0]

```

```python
# Transform the response dictionary to a DataFrame
libra_df = pd.DataFrame.from_dict(libra_headlines["articles"])

libra_df.head()

```
